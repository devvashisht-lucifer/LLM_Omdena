```python
import pandas as pd

# Load the dataset with utf-8 encoding
df = pd.read_csv(r"C:\Users\vashi\Downloads\Final_dataset2.csv", encoding='utf-8')

# Display basic information
print("Dataset Shape:", df.shape)
print("\
Columns:", df.columns.tolist())
print("\
First few rows:")
print(df.head())
print("\
Missing Values:")
print(df.isnull().sum())
print("\
Basic Statistics:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Source distribution
plt.figure(figsize=(12, 6))
source_counts = df['Source'].value_counts()
sns.barplot(x=source_counts.index, y=source_counts.values)
plt.title('Distribution of News Sources')
plt.xticks(rotation=45)
plt.show()
```

```python
# Target distribution
plt.figure(figsize=(8, 6))
target_counts = df['Target_final'].value_counts()
plt.pie(target_counts.values, labels=['Negative (0)', 'Positive (1)'], autopct='%1.1f%%')
plt.title('Distribution of Target Labels')
plt.show()
```

```python
# Top categories
print("\
Top 5 News Categories:")
print(df['Category'].value_counts().head())
```

```python
# Assuming 'df' is your DataFrame and it has a 'Category' column
top_categories = df['Category'].value_counts().head()

# Print the top categories
print("Top 5 News Categories:")
print(top_categories)

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")

# Add titles and labels
plt.title('Top 5 News Categories', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Display the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Average headline length
df['headline_length'] = df['Headlines'].str.len()
print("\
Average Headline Length:", round(df['headline_length'].mean(), 2))
```

```python
# Most common authors
print("\
Top 5 Authors:")
print(df['Author'].value_counts().head())
```

```python
# Assuming 'df' is your DataFrame and it has an 'Author' column
top_authors = df['Author'].value_counts().head()

# Print the top authors
print("Top 5 Authors:")
print(top_authors)

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=top_authors.index, y=top_authors.values, palette="coolwarm")

# Add titles and labels
plt.title('Top 5 Authors', fontsize=16)
plt.xlabel('Author', fontsize=12)
plt.ylabel('Number of Articles', fontsize=12)

# Display the plot
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between article length and target
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Target_final', y='headline_length')
plt.title('Relationship Between Article Length and Target')
plt.xlabel('Target (0: Negative, 1: Positive)')
plt.ylabel('Headline Length')
plt.show()
```

```python
# Sentiment distribution across different categories
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Category', hue='Target_final', order=df['Category'].value_counts().index)
plt.title('Sentiment Distribution Across Different Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['hour'] = df['Date'].dt.hour


plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='hour', order=range(24))
plt.title('Most Active Time Periods for News Publication')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Articles')

plt.show()
```
